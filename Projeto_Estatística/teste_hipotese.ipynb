### Testes de Hipótese

Neste projeto estarei realizando diferentes tipos de testes de hipótese com diferentes situações problema.

1- Uma fábrica anuncia que o índice de nicotina nos cigarros é em média igual a 20mg por cigarro. Um laboratório realiza 20 análises do índice. Sabe-se que o índice de nicotina se distribui normalmente com variância de 4mg². Pode se aceitar a informação do fabricante ao nível de 5%?

$H_0: \mu = 20$

$H_1: \mu > 20$

```python
import numpy as np 
import matplotlib.pyplot as plt 

mu = 20
var = 4
n = 20
Ns = 1000
X_m = []
#distribuição da média amostral

for i in range(1,Ns):
    x = np.random.normal(mu, var, n)
    X_m.append(np.mean(x))

plt.figure(figsize=(6,8))
plt.hist(x=X_m, bins=15, color="gray", density=True, rwidth=0.85)
plt.axvline(x=mu, color='r', linestyle='--', label="Média")
plt.xlabel(r'$\bar{X}$', fontsize=20)
plt.ylabel(r'P($\bar{X}$)', fontsize=20)
plt.legend()
plt.show()
```

Vemos que a distribuição da média amostral será uma normal, e é nela que vamos nos basear para testar as hipóteses.

```python
# amostra observada
X_am = [22, 19, 21, 22, 20, 18, 27, 20, 21, 19, 20, 22, 17, 20, 21, 18, 25, 16, 20, 21]
x_obs = np.mean(X_am)

alpha=95 #pois os percentis são medidos P(Z<a)
xc = np.percentile(X_m, alpha)

print(f"X crítico {xc} e  X observado {x_obs}")

if x_obs < xc:
    print("Aceitamos H0")
else:
    print("Rejeitamos H0")
```

2 - Aplicação de substância gera retardo em cobaias: Tempo de reação médio normal da população $\mu = 8s$, desvio padrão $\sigma = 2s$, amostra [9.1, 9.3, 7.2, 13.3, 10.9, 7.2, 9.9, 8, 8.6, 7.5], nível de significância $\alpha = 6$%. Houve mudança nos tempos de reação?

$H_0: \mu = 8$

$H_1: \mu \ne 8$


```python
mu = 8
var = 2
n = 10
Ns = 5000
X_m = []
#distribuição da média amostral

for i in range(1,Ns):
    x = np.random.normal(mu, var, n)
    X_m.append(np.mean(x))

plt.figure(figsize=(6,8))
plt.hist(x=X_m, bins=15, color="gray", density=True, rwidth=0.85)
plt.axvline(x=mu, color='r', linestyle='--', label="Média")
plt.xlabel(r'$\bar{X}$', fontsize=20)
plt.ylabel(r'P($\bar{X}$)', fontsize=20)
plt.legend()
plt.show()
```

```python
x_am = [9.1, 9.3, 7.2, 13.3, 10.9, 7.2, 9.9, 8, 8.6, 7.5]
x_obs = np.mean(x_am)

alpha = 3

xc1 = np.percentile(X_m, alpha)
xc2 = np.percentile(X_m, 100-alpha)

print(f"X crítico 1 = {xc1}, X crítico 2 = {xc2} e X observado {x_obs}")

if x_obs > xc1 and x_obs < xc2: 
    print("Aceitamos H0")
else: 
    print("Rejeitamos H0")
```
